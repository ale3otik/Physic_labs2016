```python
mass = [548, 1042.3, 1544.1, 2009.5, 2514.5]
mass = np.array(mass)
mass = mass*1e-03

freq = []
freq.append([0., 101.1, 203.1, 303.5, 400.0, 493.6])
freq.append([0., 127.3, 255.7, 383.7, 512.3, 644.2, 775.3, 904.7])
freq.append([0., 166.0, 331.9, 498.0, 668.1])
freq.append([0., 190.5, 381.4, 576.9, 765.])
freq.append([0., 208.5, 418.3, 629., 837.1])
freq = np.array(freq)
```

```python
us = []

import math
def plot_and_calc_u(i):
    xval = np.arange(0, len(freq[i]), 1)
    yval = freq[i]
    plt.plot(xval, yval, "+")
    plt.grid()
    coef = np.polyfit(xval[1:], yval[1:], 1)
    xval_appr = np.linspace(0, len(freq[i]), 100)
    yval_appr = [coef[0]*x_i + coef[1] for x_i in xval_appr]
    plt.plot(xval_appr, yval_appr)
    plt.show()
    u = coef[0]*2*L
    freq_appr = [coef[0] * val + coef[1] for val in xval]
    freq_appr = np.array(freq_appr)
    delta = [(freq_appr - freq[i])**2]
    sigm_freq = math.sqrt(1./len(freq[i])*np.sum(delta))
    sigm_u = math.sqrt((2.*L/(i+1e-02))*sigm_freq**2)
    print "n = " + str(i)
    print "u = " + str(u)
    print "sigm u = " + str(sigm_u) 
    us.append(u)
    return u
```

```python
for i in range(len(freq)):
    plot_and_calc_u(i)
```
